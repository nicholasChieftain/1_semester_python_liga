**Два решения задачи 1**

<img src="img/Занятие_13/два.png" width="600">

```python
#Решение задачи 1 с помощью цикла for
a = []
for i in range (0,10):
    a.append(i*2)
print(a) 
```

```python
#Решение задачи 1 с помощью генератора коллекции
a = [i*2 for i in range(0,10)]
print(a)
```
